# Descriptive Analysis

```python
import pandas as pd

# Function to calculate descriptive statistics
def get_stats(sample, field):
    mode_values = sample[field].mode().tolist()

    summary = {
        "Mean (Average)": round(sample[field].mean(), 2),
        "Median (Middle)": round(sample[field].median(), 2),
        "Mode (Most Common)": mode_values,
        "Minimum": sample[field].min(),
        "Maximum": sample[field].max(),
        "Standard Deviation": round(sample[field].std(), 2),
        "Variance": round(sample[field].var(), 2)
    }
    return summary


# Load the dataset
df = pd.read_csv("daily_internet_usage_by_age_group.csv")

# Dependent variable
column_to_analyze = "total_screen_time"

# Get descriptive statistics
project_results = get_stats(df, column_to_analyze)

# Print results
print("Basic Statistics for Internet Usage:")
for key, value in project_results.items():
    print(f"{key}: {value}")
```

# Random Sampling

```python
import pandas as pd

# Descriptive statistics function
def get_stats(sample, field):
    summary = {
        "Mean (Average)": round(sample[field].mean(), 2),
        "Median (Middle)": round(sample[field].median(), 2),
        "Mode (Most Common)": sample[field].mode().tolist(),
        "Minimum": round(sample[field].min(), 2),
        "Maximum": round(sample[field].max(), 2),
        "Standard Deviation": round(sample[field].std(), 2),
        "Variance": round(sample[field].var(), 2)
    }
    return summary


# Load dataset
df = pd.read_csv("daily_internet_usage_by_age_group.csv")

# Random sampling of size 150
random_sample_150 = df.sample(n=150, random_state=42)

# Apply descriptive statistics to dependent variable
sample_results = get_stats(random_sample_150, "total_screen_time")

# Print results
print("Descriptive Statistics for Random Sample (Size 150):")
for key, value in sample_results.items():
    print(f"{key}: {value}")
```

# Systematic Sampling

```python
import pandas as pd

# Reusable function for statistical summary
def get_stats(sample, field):
    # Get the most frequent value as a list
    mode_values = sample[field].mode().tolist()

    # Calculate core statistics and round to 2 decimal places
    summary = {
        "Mean (Average)": round(sample[field].mean(), 2),
        "Median (Middle)": round(sample[field].median(), 2),
        "Mode (Most Common)": mode_values,
        "Minimum": sample[field].min(),
        "Maximum": sample[field].max(),
        "Standard Deviation": round(sample[field].std(), 2),
        "Variance": round(sample[field].var(), 2)
    }
    return summary

# Load the internet usage dataset
df = pd.read_csv("daily_internet_usage_by_age_group.csv")

# Systematic Sampling: Selecting every 10th row from the population
# This creates a structured representative sample of 280 records
systematic_sample = df.iloc[::10]

# Applying the function to the dependent variable 'total_screen_time'
results = get_stats(systematic_sample, "total_screen_time")

# Print the final statistics for the report
print("Descriptive Statistics for Systematic Sample (Every 10th Row):")
for key, value in results.items():
    print(f"{key}: {value}")
```

# Hypothesis Testing

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Load the dataset
df = pd.read_csv("daily_internet_usage_by_age_group.csv")

# Independent and dependent variables
x = df["age"]                   # Independent variable
y = df["total_screen_time"]     # Dependent variable

# Pearson correlation test
pearson_r, p_pearson = pearsonr(x, y)

# Spearman correlation test
spearman_r, p_spearman = spearmanr(x, y)

# Display results
print("--- Hypothesis Test Results ---")
print(f"Pearson Correlation Coefficient: {round(pearson_r, 4)}")
print(f"Pearson P-Value: {round(p_pearson, 4)}")

print(f"\nSpearman Correlation Coefficient: {round(spearman_r, 4)}")
print(f"Spearman P-Value: {round(p_spearman, 4)}")

# Conclusion based on significance level (α = 0.05)
if p_pearson < 0.05 or p_spearman < 0.05:
    print("\nConclusion: There is a statistically significant relationship between age and daily internet usage (Reject H0).")
else:
    print("\nConclusion: There is no statistically significant relationship between age and daily internet usage (Fail to reject H0).")
```

# One-Sample T-test

```python
import pandas as pd
from scipy.stats import ttest_1samp

# Load the dataset
df = pd.read_csv("daily_internet_usage_by_age_group.csv")

# Assumed population mean (based on overall average)
population_mean = 8.2

# Random sample of size 150 from the dependent variable
sample = df["total_screen_time"].sample(n=150, random_state=42)

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(sample, population_mean)

# Display results
print(f"Assumed Population Mean: {population_mean}")
print(f"Sample Mean: {round(sample.mean(), 2)}")
print(f"T-Statistic: {round(t_stat, 4)}")
print(f"P-Value: {round(p_value, 4)}")

# Conclusion based on significance level (α = 0.05)
if p_value < 0.05:
    print("\nConclusion: The sample mean is significantly different from the population mean (Reject H0).")
else:
    print("\nConclusion: The sample mean is not significantly different from the population mean (Fail to reject H0).")
```
